```python
#트랜스포머 라이브러리의 버전을 출력합니다.
import transformers

print(transformers.__version__)
```

```python
#파이프 라인을 통해서 정서분석을 하고자 한다고 전달하면 그에 맞는 모델과 토큰화기를 자동으로 받을 수 있습니다.
from transformers import pipeline

senti = pipeline('sentiment-analysis')
```

```python
# 그럼 어떤 모델을 다운받았는지 확인해봅니다.
# 출력문을 보면 받은 모델은 DistilBert라는 것을 확인할 수 있습니다.
senti.model
```

```python
# 모델에 곧바로 문장을 넣어서 정서분석 결과값을 받습니다.
print(senti(["There is but only one truly serious philosophical problem. It is suicide."]))
```

```python
print(senti(["It is a truth universally acknowledged that a single man in possession of a good fortune must be in want of a wife."]))
```

```python
print(senti(["Call me Ishmael."]))
```
