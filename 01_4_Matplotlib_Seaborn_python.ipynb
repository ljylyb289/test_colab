# Scatterplot - adding bells and whistles
The plot above is OK but looks barebone. We can add few bells and whistles to it with simple code.

* Figure size
* Main title (and size the font)
* X- and Y-labels (and size the font)
* X- and Y-ticks (and size the font)
* Grid
* Text for labelling
* Legend
* Vertical (or horizontal) lines

```python
# 8X6인치인 새로운 그림 상자를 만들어냅니다.
plt.figure(figsize=(8,6))

# 제목을 설정한것으로 폰트는 이탤릭, 사이즈는 20입니다.
plt.title("Plot of Age vs. Height (in cms)\n",fontsize=20, fontstyle='italic')

# X- and Y-label with fontsize
plt.xlabel("Age (years)",fontsize=16)
plt.ylabel("Height (cms)",fontsize=16)

# 격자선옵션에 True를 입력해서 격자선을 넣었습니다.
plt.grid (True) 

# y축의 범위를 100에서 200까지로 설정했습니다
plt.ylim(100,200)

# ticks함수는 각 축의 눈금의 간격과 폰트 사이즈를 지정할수 있습니다. x축에서는 0,5,10와 같이 5의 간격으로 55까지 나타납니다.
plt.xticks([i*5 for i in range(12)],fontsize=15)
plt.yticks(fontsize=15)

# Main plotting function with choice of color, marker size, and marker edge color
plt.scatter(x=age,y=height,c='orange',s=150,edgecolors='k')

# Adding bit of text to the plot
plt.text(x=15,y=105,s="Height increaes up to around \n20 years and then tapers off",fontsize=15, 
         rotation=30, linespacing=2)
plt.text(x=22,y=185,s="Nobody has a height beyond 180 cm",fontsize=15)

# 이는 세로선을 긋는 함수이며 x좌표 20에서 y축으로 100~180까지 그으며 점선이고 너비가 3에 파란색이라는 것까지 정했습니다.
plt.vlines(x=20,ymin=100,ymax=180,linestyles='dashed',color='blue',lw=3)

# 이는 가로선을 긋는 함수이며  y값은 180에 위치하고 x의 범위는 0~55이다. 점선이며 너비가 3이고 색깔은 붉은색입니다.
plt.hlines(y=180,xmin=0,xmax=55,linestyles='dashed',color='red',lw=3)

# Adding a legend
plt.legend(['Height in cms'],loc=2,fontsize=14)

# 이렇게 구현된 플롯을 실제로 그려서 표현해줍니다.
plt.show()
```

# Bar chart

x= x축에 들어가는 데이터


height= y축에 들어가는 데이터


width= 각 막대의 너비


color= 막대의 색깔


edgecolor= 막대의 테두리선 색깔


alpha= 불투명도

막대 색깔을 검은색으로 만들고 테두리는 빨간색으로 만들었으며 불투명도는 0.9로 올렸습니다. 

```python
plt.figure(figsize=(12,4))
plt.title("people's weight",fontsize=30, fontstyle='italic')
# Main plot function 'bar'
plt.bar(x=people,height=weight, width=0.6,color='black',edgecolor='red',alpha=0.9)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

# Histogram

**[숙제5-1]**  
figsize는 7x5인치의 빈 그림 박스를 생성한다는 의미입니다.
plit.hist에서 weight는 이 데이터를로 히스토그램을 만든다는 의미입니다. color는 이 히스토그램의 색깔을 설정하고 edgecolor는 테두리 색깔을 의미합니다. alpha는 불투명도와 bins는 구간을 몇개로 나눌지 설정합니다.
plt.xticks의 fontsize는 눈금의 폰트 사이즈를 설정합니다.

막대의 색깔은 검은색으로 바꾸었으며 구간의 개수를 3개로 축소시키고 불투명도를 0.2로 만들었습니다.

```python
import numpy as np

plt.figure(figsize=(10,10))

# Main plot function 'hist'

plt.hist(weight,color='black',edgecolor='k', alpha=0.2,bins=3)


plt.title("Histogram of patient weight",fontsize=40)
plt.xlabel("Weight in kgs",fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

# Simple line plot

dayys라는 변수에 1부터 30까지의 숫자를 넣는 배열을 생성합니다.
그다음에 np.random.randn(30)을 통해서 무작위로 30개의 값을 정규분포를 사용해서 생성합니다. 그 값을 주어진 식을 통해서 각 days값에 연산을 시행합니다.

```python
days = np.arange(1,31)
candidate_A = 50+days*0.07+2*np.random.randn(30)
candidate_B = 50-days*0.1+3*np.random.randn(30)
```

candidate_A와 ncandidate_B에서 각각 최대값과 최소값을 구하고 그 최대,최소값끼리 비교해서 전체에서 최대인 값과 최소인 값을 구합니다.

```python
# Determine the minimum and maximum of stock prices
ymin = min(candidate_A.min(),candidate_B.min())
ymax = max(candidate_A.max(),candidate_B.max())
```

 plt.style.use('fivethirtyeight')는 fivethirtyeight이라는 스타일을 가져와서 쓰겠다는 의미입니다.
plt.plot(days,candidate_A,'o-',markersize=10,c='blue',lw=2)
days는 x축 데이터를 의미하며 candidate_Asms y축 데이터를 의미합니다. o-는 마커를 원으로 지정하고 이들을 선으로 연결하라는 의미입니다. markersize는 마커의 사이즈를 지정합니다.c는 마커와 선의 색깔을 지정하며 lw는 선폭의 의미합니다.


A측의 속성은 색깔을 검은색으로 모양을 삼각형으로 바꾸고 선의 굵기도 4로 바꾸었습니다. B측은 빨간색에 선읜 굵기는 1로 바꾸고 모양은 원형으로 만들었습니다.


```python
# Set style
plt.style.use('fivethirtyeight')

plt.figure(figsize=(12,5))
plt.title("Time series plot of poll percentage over a month\n",fontsize=20, fontstyle='italic')
plt.xlabel("Days",fontsize=16)
plt.ylabel("Poll percentage (%)",fontsize=16)
plt.grid (True)
plt.ylim(ymin*0.98,ymax*1.02)
plt.xticks([i*3 for i in range(16)],fontsize=14)
plt.yticks(fontsize=15)

# Main plotting function - plot (note markersize, lw (linewidth) arguments)
plt.plot(days,candidate_A,'^-',markersize=10,c='black',lw=4)
plt.plot(days,candidate_B,'o-',markersize=10,c='red',lw=1)

plt.legend(['Poll percentage of candidate A (%)', 'Poll percentage of candidate B (%)'],loc=2,fontsize=14)
plt.show()
```

x는 x축의 값을 candidate_A,candidate_B 2개를 놓는다는 것을 의미합니다. showmeans는 True로 놓을경우 평균값을 마커로 표시해줍니다.

```python
plt.style.use('ggplot')


# Note how to convert default numerical x-axis ticks to the list of string by passing two lists 
plt.boxplot(x=[candidate_A,candidate_B],showmeans=True)
plt.grid(True)
plt.xticks([1,2],['Candidate A','Candidate B'])

#plt.yticks(fontsize=15)

plt.show()
```

# Seaborn - advanced statistical visualizations

```python
import seaborn as sns
```

# Boxplot separated by class/groups of data

가운데 박스에서 그어진 선은 중앙값을 의미합니다. 박스는 25%지점을 아랫변, 75%를 윗변으로 둡니다. 박스를 벗어난 가로선은 윗부분이 데이터의 최대값, 아래가 최솟값입니다. 이 최대 최소를 벗어난 점은 특이한 분포를 가지는 데이터로 봅니다. 

```python
sns.boxplot(x='Class',y='Alcohol',data=df)
plt.show()
```

# Violin plots (combination of boxplot and histogram/kernel density)


sns.violinplot(x='Class',y='Alcohol',data=df)x는 x축에 들어가는 데이터가 무엇인지 y는 y축에 들어가는 데이터가 무엇인지 설정합니다. data=df는 df라는 데이터 셋에서 가져온 데이터라는 의미입니다.

x축 y축이 서로 뒤바뀌며 이전의 그래프와는 사뭇 다른 모양으로 나옵니다.

```python
sns.violinplot(x='Alcohol',y='Class',data=df)
plt.show()
```

# regplot - computes and plots the linear regression fit along with confidence interval

```python
sns.regplot(x='Alcohol',y='Color intensity',data=df)
plt.show()
```

# lmplot - combination of regplot with grid to visualize across various groups/classes

```python
sns.lmplot(x='Alcohol',y='Color intensity',hue='Class',data=df)
plt.show()
```

# Correlation matrix and heatmap

```python
import numpy as np
```

corrcoef는 상관계수를 구해주는 함수입니다. 다만 주어진 코드에서는 df.rowvar를 False로 둠으로서 행이 기준이아닌 열을 기준으로 상관계수를 구했습니다.

```python
corr_mat=np.corrcoef(df,rowvar=False)
```

corr.mat의 행과 열의 크기를 구해줍니다.

```python
corr_mat.shape
```

새로운 데이터 프레임을 기존에 연산한 상관계수를 원소로 해서 만듭니다. 열과 인덱스는 기존의 데이터의 열에서 가져옵니다.

```python
corr_df = pd.DataFrame(corr_mat,columns=df.columns,index=df.columns)
```

round함수는 corr.mat에서 데이터들을 소수점 3자리까지 반올림 하도록 만듭니다.

```python
print(np.round(corr_mat,3))
```

cmap을 YlGnBu로 바꾸고 linewidth를 2로 바꾸었습니다.

```python
sns.heatmap(corr_df,linewidth=2,cmap='YlGnBu')
plt.show()
```
